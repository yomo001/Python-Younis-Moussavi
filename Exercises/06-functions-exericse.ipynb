## 2. Euclidean distance (*)

The formula for Euclidean distance in 2D between $P: (p_1, p_2)$ and $Q: (q_1, q_2)$  is: 

$d(P,Q) = \sqrt{(p_1-q_1)^2 + (p_2-q_2)^2}$

&nbsp; a) &nbsp; Create a function that takes two points as input parameters and return the Euclidean between them. (*)

&nbsp; b) &nbsp; Let the user input two points. Call the function using the users input points. (*)

&nbsp; c) &nbsp; Use your function to calculate distances between the origin (0, 0) and each of these points: (10, 3), (-1, -9), (10, -10), (4, -2), (9, -10). (*)

<details>

<summary>Hint</summary>

Use a **for** statement

</details>
<br>
<details>

<summary>Answer</summary>

c)

```
[10.4, 9.1, 14.1, 4.5, 13.5]
```

</details>


````python
# Solution with foor loop
import math

#a)
def euclidean_distance(x1,y1,x2,y2):
    return f"({math.sqrt((x2-x1)**2 + (y2-y1)**2):.2f})"            # 1 decimal, also with f string possible f"{distance_list:.1f}"

#b)
for _ in range(1):
    x1 = float(input("Please enter the x-value for point 1: "))
    y1 = float(input("Please enter the y-value for point 1: "))
    x2 = float(input("Please enter the x-value for point 2: "))
    y2 = float(input("Please enter the y-value for point 2: "))
euclidean_distance(x1, y1, x2, y2)
````

````python
#c) # Solution with foor loop

origin = (0,0)
points_list = [(10, 3), (-1, -9), (10, -10), (4, -2), (9, -10)]

distance_list = []

for point in points_list:
    x1, y1 = point
    x2, y2 = origin
    distance_list.append(euclidean_distance(x1, y1, x2, y2))

print(f"{distance_list}")
````

````python
#c) # Solution with comprehension

origin = (0,0)
points_list = [(10, 3), (-1, -9), (10, -10), (4, -2), (9, -10)]

distance_list = [euclidean_distance(x, y, origin[0], origin[1]) for x, y in points_list]


print(f"{distance_list}")
````

Make the following functions with **def** or **lambda** and plot their graphs in the same figure window, with $x\in [-10,10]$ :

&nbsp; a) &nbsp;  $f(x) = x^2 -3$ (*)

&nbsp; b) &nbsp; $g(x) = 4x-7$ (*)

What could the relation between $f(x)$ and $g(x)$ be?

````python
import numpy as np
import matplotlib.pyplot as plt

f = lambda x: x**2 - 3
g = lambda x: 4*x - 7

x = np.linspace(-10,10,400)

y_f = f(x)
g_f = g(x)

# Plot the graphs of f(x) and g(x) in the same figure
plt.figure(figsize=(8, 6))
plt.plot(x, y_f, label = "f(x) = x**2 - 3", color = "blue")
plt.plot(x, g_f, label = "g(x) = 4*x - 7", color = "red")

# Add labels and a legend
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graphs of f(x) and g(x)')
plt.legend()

# Show the plot
plt.grid(True)
plt.axhline(y=0, color='black', linewidth=0.7)
plt.axvline(x=0, color='black', linewidth=0.7)
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Define the functions using def
def f(x):
    return x**2 - 3

def g(x):
    return 4*x - 7

# Create an array of x values
x = np.linspace(-10, 10, 400)

# Calculate the corresponding y values for f(x) and g(x)
y_f = [f(val) for val in x]
y_g = [g(val) for val in x]

# Plot the graphs of f(x) and g(x) in the same figure
plt.figure(figsize=(8, 6))
plt.plot(x, y_f, label='f(x) = x^2 - 3', color='blue')
plt.plot(x, y_g, label='g(x) = 4x - 7', color='red')

# Add labels and a legend
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graphs of f(x) and g(x)')
plt.legend()

# Show the plot
plt.grid(True)
plt.axhline(y=0, color='black', linewidth=0.7)
plt.axvline(x=0, color='black', linewidth=0.7)
plt.show()
````

## 4. Name cleaner (*)

Create a function that takes a name as an input and:

- removes all leading and trailing blank spaces
- make capitalize the first character of each name, and make the rest lowercase

Use your function on this list of strings: 

```
["  MaRcUs ", " iDA aNderSon", "OLOF Olofsson            "  ]
```

````python
def name_cleaner(name = False):
    if name:
        return name.strip().title()
    else:
        name = input("Please enter your name: ")
        return name.strip().title()

name_list = ["  MaRcUs ", " iDA aNderSon", "OLOF Olofsson            "  ]
for name in name_list:
    print(name_cleaner(name))
````

## 5. Change (**)

Create a function that takes a value as input parameter and print out the banknotes and coins in Swedish currency representing this value. For example 5289 would give the following printout: 

- 5st 1000-lapp
- 1st 200-lapp
- 1st 50-lapp
- 1st 20-lapp
- 1st 10-krona
- 1st 5-krona
- 2st 2-krona

Now let the user input a value, and use the function to calculate the change.

````python
def currency_changer(amount):
    tusenlappar = amount // 1000
    amount -= tusenlappar * 1000
    tvåhundralappar = amount // 200
    amount -= tvåhundralappar * 200
    femtiolappar = amount // 50
    amount -= femtiolappar *50
    tjugolappar = amount // 20
    amount -= tjugolappar * 20
    tior = amount // 10
    amount -= tior * 10
    femmor = amount // 5
    amount -= femmor * 5
    tvåor = amount // 2 
    amount -= tvåor * 2
    print(f"""                                       
          - {tusenlappar}st 1000-lapp
          - {tvåhundralappar}st 200-lapp
          - {femtiolappar}st 50-lapp
          - {tjugolappar}st 20-lapp
          - {tior}st 10-krona
          - {femmor}st 5-krona
          - {tvåor}st 2-krona""")
# Med triple quotes behöver inte \n anges för newlines, med double quotes får varje rad ha doublequotes och \n


currency_changer(13242)
````
